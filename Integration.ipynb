```python
import numpy as np
import matplotlib.pyplot as plt

# Function definition
def f(x):
    return x**2

# Left Hand Riemann Sum
def left_riemann_sum(a, b, delta_x):
    x = np.arange(a, b, delta_x)
    return sum(f(x) * delta_x)

# Right Hand Riemann Sum
def right_riemann_sum(a, b, delta_x):
    x = np.arange(a + delta_x, b + delta_x, delta_x)
    return sum(f(x) * delta_x)

# Midpoint Method
def midpoint(a, b, delta_x):
    x = np.arange(a + 0.5 * delta_x, b, delta_x)
    return sum(f(x) * delta_x)

# Trapezoidal Rule
def trapezoidal(a, b, delta_x):
    x = np.arange(a, b + delta_x, delta_x)
    y = f(x)
    return delta_x * (0.5 * y[0] + 0.5 * y[-1] + sum(y[1:-1]))

a, b = 0, 4
delta_xs = [10**(-i) for i in reversed(range(5))]
actual_value = (b**3)/3 - (a**3)/3  # Integral of x^2 from 0 to 4

left_values, right_values, mid_values, trap_values = [], [], [], []

for delta_x in delta_xs:
    left_values.append(left_riemann_sum(a, b, delta_x))
    right_values.append(right_riemann_sum(a, b, delta_x))
    mid_values.append(midpoint(a, b, delta_x))
    trap_values.append(trapezoidal(a, b, delta_x))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(delta_xs, left_values, label="Left Riemann Sum", marker='o')
plt.plot(delta_xs, right_values, label="Right Riemann Sum", marker='x')
plt.plot(delta_xs, mid_values, label="Midpoint Method", marker='.')
plt.plot(delta_xs, trap_values, label="Trapezoidal Rule", marker='s')
plt.axhline(y=actual_value, color='r', linestyle='-', label="Actual Value")
plt.xscale('log')
plt.xlabel("x")
plt.ylabel("Approximated Value")
plt.title("Approximation of x^2 from 0 to 4")
plt.legend()
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - x - 1

def bisection_method(a, b, tol, max_iter):
    print("Bisection Method Values : ")
    x_vals = []
    approx_errors = []
    relative_approx_errors = []

    for i in range(max_iter):
        c = (a + b) / 2
        x_vals.append(c)


        if i > 0:
            approx_error = abs(x_vals[i] - x_vals[i - 1])
            approx_errors.append(approx_error)
            relative_approx_error = approx_error / abs(x_vals[i])
            relative_approx_errors.append(relative_approx_error)
            # print("a = " ,a, "b = " ,b , "c = ",c, "app_err = ",approx_error,"rel_app_err = ",relative_approx_error)

            # print(c)

            if approx_error < tol:
                break

        if f(c) == 0 :
            break
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c

    return x_vals, approx_errors, relative_approx_errors

def false_position_method(a, b, tol, max_iter):
    print("False Position Method Values : ")
    x_vals = []
    approx_errors = []
    relative_approx_errors = []

    for i in range(max_iter):
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        x_vals.append(c)

        if i > 0:
            approx_error = abs(x_vals[i] - x_vals[i - 1])
            approx_errors.append(approx_error)
            relative_approx_error = approx_error / abs(x_vals[i])
            relative_approx_errors.append(relative_approx_error)
            # print("a = " ,a, "b = " ,b , "c = ",c, "app_err = ",approx_error,"rel_app_err = ",relative_approx_error)

            if approx_error < tol:
                break

        if f(c) == 0 :
            break
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c

    return x_vals, approx_errors, relative_approx_errors

# Initial values
x0 = 50
x1 = -50
tolerance = 1e-6
max_iterations = 1000



# Bisection method
bisection_x_vals, bisection_approx_errors, bisection_relative_approx_errors = bisection_method(x0, x1, tolerance, max_iterations)

# False Position method
false_position_x_vals, false_position_approx_errors, false_position_relative_approx_errors = false_position_method(x0, x1, tolerance, max_iterations)

# Bar Chart
iteration_numbers_bisec = np.arange(1, len(bisection_approx_errors)+1)
iteration_numbers_false = np.arange(1, len(false_position_approx_errors)+1)

# print(len(iteration_numbers),len(bisection_approx_errors),len(bisection_relative_approx_errors))
# print(len(iteration_numbers),len(false_position_approx_errors),len(false_position_relative_approx_errors))



plt.bar(iteration_numbers_bisec, bisection_approx_errors, color='blue', label='Bisection')
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Approximation Error vs Iteration Number(Bisection)')
plt.legend()
plt.show()


plt.bar(iteration_numbers_bisec, bisection_relative_approx_errors, color='blue', label='Bisection')
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Relative Approximation Error vs Iteration Number(Bisection)')
plt.legend()
plt.show()


plt.bar(iteration_numbers_false, false_position_approx_errors, color='orange', label='False Position')
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Approximation Error vs Iteration Number(False)')
plt.legend()
plt.show()



plt.bar(iteration_numbers_false, false_position_relative_approx_errors, color='orange', label='False Position')
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Relative Approximation Error vs Iteration Number(False)')
plt.legend()
plt.show()

# Line Chart for Relative Approximation Errors
plt.plot(iteration_numbers_bisec, bisection_approx_errors, marker='o', color='blue', label='Bisection')
plt.plot(iteration_numbers_false, false_position_approx_errors, marker='o', color='orange', label='False Position')
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Approximation Error vs Iteration Number')
plt.legend()
plt.show()


# Line Chart for Relative Approximation Errors
plt.plot(iteration_numbers_bisec, bisection_relative_approx_errors, marker='o', color='blue', label='Bisection')
plt.plot(iteration_numbers_false, false_position_relative_approx_errors, marker='o', color='orange', label='False Position')
plt.xlabel('Iteration Number')
plt.ylabel('Relative Approximation Error')
plt.title('Relative Approximation Error vs Iteration Number')
plt.legend()
plt.show()

```

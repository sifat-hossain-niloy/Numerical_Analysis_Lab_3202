```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - x - 1

def secant_method(x0, x1, tol=1e-6, max_iter=100):
    app_err = []
    rel_app_err = []
    x_values = []
    i = 0

    while i < max_iter:
        xi_minus_1 = x0
        fi_minus_1 = f(xi_minus_1)

        xi = x1
        fi = f(xi)

        if xi == xi_minus_1:
            print("xi and xi_minus_1 are the same. Secant method may not converge.")
            return None

        f_prime_i = (fi - fi_minus_1) / (xi - xi_minus_1)

        xi_plus_1 = xi - fi / f_prime_i
        approx_error = abs(xi_plus_1 - xi)
        rel_approx_error = abs(approx_error / xi_plus_1)

        print(f"Iteration {i+1}:")
        print(f"xi-1 = {xi_minus_1:.7f}, f(xi-1) = {fi_minus_1:.7f}")
        print(f"xi = {xi:.7f}, f(xi) = {fi:.7f}, f'(xi) = {f_prime_i:.7f}")
        print(f"xi+1 = {xi_plus_1:.7f}, Approximation error = {approx_error:.7f}, Relative approximation error = {rel_approx_error:.7f}")

        x_values.append(xi_plus_1)
        app_err.append(approx_error)
        rel_app_err.append(rel_approx_error)

        if approx_error < tol:
            print("Converged successfully!")
            break

        x0 = xi
        x1 = xi_plus_1
        i += 1

    return x_values, app_err, rel_app_err

x0_secant = 50
x1_secant = 48

x_values_secant, app_err_secant, rel_app_err_secant = secant_method(x0_secant, x1_secant)

# approximation errors
iterations_secant = list(range(1, len(app_err_secant) + 1))
plt.bar(iterations_secant, app_err_secant)
plt.xlabel('Iteration Number')
plt.ylabel('Approximation Error')
plt.title('Approximation Errors using Secant Method')
plt.show()

# relative approximation error
plt.bar(iterations_secant, rel_app_err_secant)
plt.xlabel('Iteration Number')
plt.ylabel('Relative Approximation Error')
plt.title('Relative Approximation Errors using Secant Method')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def left_hand_riemann_sum(a, b, n):
    dx = (b - a) / n
    return sum(f(a + i*dx) for i in range(n)) * dx

def right_hand_riemann_sum(a, b, n):
    dx = (b - a) / n
    return sum(f(a + i*dx) for i in range(1, n+1)) * dx

def midpoint_method(a, b, n):
    dx = (b - a) / n
    return sum(f(a + (i + 0.5)*dx) for i in range(n)) * dx

def trapezoidal_rule(a, b, n):
    dx = (b - a) / n
    return (f(a) + f(b) + 2*sum(f(a + i*dx) for i in range(1, n))) * dx / 2

def simpsons_rule(a, b, n):
    dx = (b - a) / n
    return (f(a) + f(b) + 4*sum(f(a + (i + 0.5)*dx) for i in range(n)) + 2*sum(f(a + i*dx) for i in range(1, n))) * dx / 6


a, b = 0, 2
actual_value = (b**3)/3 - (a**3)/3
n_values = [i for i in range(1, 10001)]  # Varying n from 1 to 100,00
left_values, right_values, mid_values, trap_values, simp_values = [], [], [], [], []

for n in n_values:
    left_values.append(left_hand_riemann_sum(a, b, n))
    right_values.append(right_hand_riemann_sum(a, b, n))
    mid_values.append(midpoint_method(a, b, n))
    trap_values.append(trapezoidal_rule(a, b, n))
    simp_values.append(simpsons_rule(a, b, n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, [actual_value]*len(n_values), label="Actual Value", linestyle="--")
plt.plot(n_values, left_values, label="Left Riemann Sum")
plt.plot(n_values, right_values, label="Right Riemann Sum")
plt.plot(n_values, mid_values, label="Midpoint Method")
plt.plot(n_values, trap_values, label="Trapezoidal Rule")
plt.plot(n_values, simp_values, label="Simpson's Rule")
plt.xlabel('n')
plt.ylabel('Approximated Value')
plt.legend('Approximated value vs. Actual value')
plt.show()
```

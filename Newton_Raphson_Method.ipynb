```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**3 - x - 1

def df(x):
    return 3*x**2 - 1

# Newton-Raphson method
def newton_raphson(x0, tol=1e-6, max_iter=100):
    results = []
    x = x0

    for i in range(max_iter):
        x_new = x - f(x)/df(x)

        # Compute errors
        approx_error = abs(x_new - x)
        rel_approx_error = approx_error / abs(x_new) if x_new != 0 else None

        # Store iteration details
        results.append((x, f(x), df(x), x_new, approx_error, rel_approx_error))

        if approx_error < tol:
            break

        x = x_new

    return results

# Start from x0 = 50
x0 = 50
iterations = newton_raphson(x0)

# Print results
print("xi\t\tf(xi)\t\tf'(xi)\t\txi+1\t\tApprox. Error\tRel. Approx. Error")
for (x, fx, dfx, x_new, approx_error, rel_approx_error) in iterations:
    print(f"{x:.6f}\t{fx:.6f}\t{dfx:.6f}\t{x_new:.6f}\t{approx_error:.6f}\t{rel_approx_error:.6f}")

# Plotting
approx_errors = [res[4] for res in iterations]
iterations_nums = list(range(1, len(approx_errors)+1))
plt.bar(iterations_nums, approx_errors)
plt.xlabel("Iteration Number")
plt.ylabel("Approximation Error")
plt.title("Approximation Error vs Iteration Number")
plt.show()
```
